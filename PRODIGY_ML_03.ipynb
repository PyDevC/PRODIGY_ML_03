```python
import os
from skimage.io import imread
from skimage.transform import resize
```

```python
train_images = []
labels = []
test_images = []

dir = '/kaggle/input/cats-and-dogs-image-classification/'
for d in os.listdir(os.path.join(dir)):
    if d == 'train':
        for da in os.listdir(os.path.join(dir,d)):
            for img in os.listdir(os.path.join(dir,d,da)):
                train_images.append(os.path.join(dir,d,da,img))
                if da == 'dogs':
                    labels.append(1)
                else:
                    labels.append(0)
    else:
        for da in os.listdir(os.path.join(dir,d)):
            for img in os.listdir(os.path.join(dir,d,da)):
                test_images.append(os.path.join(dir,d,da,img))
                
```

```python
Iseries = pd.Series(train_images,name = "Image")
Lseries = pd.Series(labels, name = "Label")
data = pd.concat([Iseries,Lseries],axis = 1)
df = pd.DataFrame(data)

test_Iseries = pd.Series(test_images, name = "Images")
test = pd.DataFrame(test_Iseries)
```

```python
from cv2 import imread
from cv2 import resize
from sklearn.decomposition import PCA
from sklearn import svm
from sklearn.svm import SVC
```

```python
def transformation(path):
    img = imread(path)
    img = resize(img,(64, 64))
    return img
```

```python
import PIL
from PIL import Image

x = []
y = df['Label']

for dir in df['Image']:
    x.append(transformation(dir))

X = np.array(x)
y = np.array(y)
```
